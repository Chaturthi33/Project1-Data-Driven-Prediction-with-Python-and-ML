# Implementation

### KNN Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix,classification_report
```

```python
# Exchange object to numeric
encoder = LabelEncoder()
df['Kraljic_Category'] = encoder.fit_transform(df['Kraljic_Category'])
```

```python
df
```

```python
# Select features (X) and target (y)
X = df[[
    "Lead_Time_Days", "Order_Volume_Units", "Cost_per_Unit",
    "Supply_Risk_Score", "Profit_Impact_Score", "Environmental_Impact"]]
y = df["Kraljic_Category"]
```

```python
# Split Dataset
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.2,random_state=42)
```

```python
# Feature Scaling
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
# Create and fit KNN model
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled,y_train)
```

```python
# Predication
y_pred=knn.predict(X_test_scaled)
```

```python
# Evalutation
print("Accuracy:",accuracy_score(y_test,y_pred))
print("\n Confusion Matrix:\n", confusion_matrix(y_test,y_pred))
print("\n Classification Report:\n",classification_report(y_test,y_pred))
```

### Insight
.Good pipeline: preprocessing → training → evaluation.

.Using only 20% training data, so try train_size=0.8 instead.

.This code builds a KNN classifier to predict the Kraljic Category of products. It evaluates performance using accuracy, confusion matrix, and                 classification report. The main issue is that training set is too small, which may cause underfitting.

.Overall Accuracy: 97.75%

.Excellent performance — the model predicts the correct Kraljic Category almost all the time.

.Means KNN with K=5 works very well for this dataset.
